```python
import numpy as np
import matplotlib.pyplot as plt

N = 32
k = 1
n = np.arange(N)

# Definir x(n) como una funcin seno (por ejemplo, seno de una frecuencia determinada)
f = 1  # Frecuencia de la seal seno (puedes ajustarla)
x = np.sin(2 * np.pi * f * n / N)  # Funcin seno

# Inicializar la sumatoria
S = np.zeros(N, dtype=complex)

# Usar un bucle para calcular la sumatoria con += para cada n
for i in range(N):
    S[i] = np.sum(x[i] * np.exp(-1j * k * 2 * np.pi * n / N))

# Calcular el mdulo (valor absoluto) de la sumatoria
mod_S = np.abs(S)

fase_S = np.angle(S)

# Graficar la parte real de la sumatoria
plt.figure(figsize=(10, 6))
plt.stem(n, np.real(S), label='Parte real de S')
plt.title(f"Parte real de la sumatoria S para k={k}")
plt.xlabel("n")
plt.ylabel("Re(S)")
plt.grid(True)
plt.legend()
plt.show()

# Graficar la parte imaginaria de la sumatoria
plt.figure(figsize=(10, 6))
plt.stem(n, np.imag(S), label='Parte imaginaria de S', linefmt='r-', markerfmt='ro')
plt.title(f"Parte imaginaria de la sumatoria S para k={k}")
plt.xlabel("n")
plt.ylabel("Im(S)")
plt.grid(True)
plt.legend()
plt.show()

# Graficar el mdulo de la sumatoria
plt.figure(figsize=(10, 6))
plt.stem(n, mod_S, label='Mdulo de S', linefmt='g-', markerfmt='go')
plt.title(f"Mdulo de la sumatoria S para k={k}")
plt.xlabel("n")
plt.ylabel("|S|")
plt.grid(True)
plt.legend()
plt.show()

# Graficar la fase de la sumatoria (en radianes)
plt.figure(figsize=(10, 6))
plt.stem(n, fase_S, label='Fase de S', linefmt='b-', markerfmt='bo')
plt.title(f"Fase de la sumatoria S para k={k}")
plt.xlabel("n")
plt.ylabel("Fase(S) [radianes]")
plt.grid(True)
plt.legend()
plt.show()

```
