## Introducción

En sistemas **discretos LTI**, la relación entre la entrada $x(n)$ y la salida $y(n)$ puede escribirse como una **ecuación en diferencias**.  
Al aplicar la **Transformada Z**, se obtiene una relación algebraica:

$$
T(z) = \frac{Y(z)}{X(z)}
$$

donde $T(z)$ es la **función de transferencia** del sistema. Describe cómo cada componente frecuencial de la entrada se modifica en amplitud y fase.

En los **filtros FIR**, $T(z)$ es un polinomio en potencias negativas de $z$ (solo términos de entrada); por eso son **no recursivos** y **estables**, y suelen presentar **fase casi lineal**.

Para estudiar el comportamiento espectral se evalúa $T(z)$ en el **círculo unidad**:

$$
z = e^{j\omega}
$$

y así se obtiene la **respuesta en frecuencia**:

$$
H(e^{j\omega}) = T(z)\bigg|_{z=e^{j\omega}}
$$

- **Módulo:** $|H(e^{j\omega})|$ indica la ganancia/atenuación por frecuencia.  
- **Fase:** $\angle H(e^{j\omega})$ indica el desplazamiento de fase.

En general, los **promedios móviles** atenúan altas frecuencias (pasa-bajos) y las **diferencias** realzan cambios rápidos (pasa-altos).

**Análisis de resultados – Punto 1**

- Los **sistemas (a)** y **(b)** son filtros de **media móvil**.  
  Promedian varias muestras consecutivas de la señal de entrada, reduciendo las variaciones rápidas y el ruido.  
  Por eso actúan como **filtros pasa-bajos**, que dejan pasar las bajas frecuencias y atenúan las altas.

- Los **sistemas (c)** y **(d)** son **filtros diferencia**.  
  Calculan la diferencia entre una muestra y otra separada por uno o dos períodos.  
  Esto resalta los cambios o transiciones rápidas en la señal, eliminando las partes constantes, por lo que se comportan como **filtros pasa-altos**.

**Conclusión:**

Los cuatro sistemas son **FIR (Finite Impulse Response)**, es decir, tienen una duración finita de respuesta al impulso y no utilizan realimentación.  
Por esta razón, son **incondicionalmente estables** y de fácil implementación.  
Mientras que los de media móvil suavizan la señal, los de diferencia permiten detectar variaciones bruscas o bordes en la señal de entrada.

### Punto 2 – Respuesta en frecuencia $$H(e^{j\omega})$$

Para estudiar el comportamiento espectral de cada sistema, se evalúa la función de transferencia sobre el **círculo unidad** del plano Z, reemplazando:

$$
z = e^{j\omega}
$$

De esta forma, se obtiene la **respuesta en frecuencia** del sistema:

$$
H(e^{j\omega}) = T(z)\big|_{z=e^{j\omega}}
$$

**Análisis de resultados – Punto 2**

- **Sistemas (a) y (b) – Media móvil (pasa-bajos)**
  
  El módulo presenta un máximo en
  $$
  \omega = 0,
  $$
  y disminuye a medida que crece la frecuencia. Esto indica que se **preservan las bajas frecuencias** y se **atenuan las altas**.  
  Además aparecen ceros periódicos sobre el círculo unidad, generando frecuencias donde
  $$
  |H(e^{j\omega})| = 0.
  $$

- **Sistemas (c) y (d) – Diferencia (pasa-altos)**
  
  El módulo es nulo en
  $$
  \omega = 0
  $$
  y crece hasta cercanías de
  $$
  \omega = \pi,
  $$
  lo que refleja que **se cancelan componentes lentas/constantes** y **se realzan variaciones rápidas**.

- **Fase en todos los sistemas**
  
  La fase es aproximadamente lineal con la frecuencia, por lo que el sistema introduce **retardo casi constante** (sin distorsión de fase significativa).

**Conclusión:**

- (a) y (b): filtros de **media móvil** ⇒ comportamiento **pasa-bajos**.  
- (c) y (d): filtros de **diferencia** ⇒ comportamiento **pasa-altos**.  
- En todos los casos, la fase **lineal** implica un desplazamiento temporal consistente sin deformar la forma de onda.
