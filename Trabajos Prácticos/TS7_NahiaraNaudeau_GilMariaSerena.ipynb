## Introducción

En sistemas **discretos LTI**, la relación entre la entrada $x(n)$ y la salida $y(n)$ puede escribirse como una **ecuación en diferencias**.  
Al aplicar la **Transformada Z**, se obtiene una relación algebraica:

$$
T(z) = \frac{Y(z)}{X(z)}
$$

donde $T(z)$ es la **función de transferencia** del sistema. Describe cómo cada componente frecuencial de la entrada se modifica en amplitud y fase.

En los **filtros FIR**, $T(z)$ es un polinomio en potencias negativas de $z$ (solo términos de entrada); por eso son **no recursivos** y **estables**, y suelen presentar **fase casi lineal**.

Para estudiar el comportamiento espectral se evalúa $T(z)$ en el **círculo unidad**:

$$
z = e^{j\omega}
$$

y así se obtiene la **respuesta en frecuencia**:

$$
H(e^{j\omega}) = T(z)\bigg|_{z=e^{j\omega}}
$$

- **Módulo:** $|H(e^{j\omega})|$ indica la ganancia/atenuación por frecuencia.  
- **Fase:** $\angle H(e^{j\omega})$ indica el desplazamiento de fase.

En general, los **promedios móviles** atenúan altas frecuencias (pasa-bajos) y las **diferencias** realzan cambios rápidos (pasa-altos).
