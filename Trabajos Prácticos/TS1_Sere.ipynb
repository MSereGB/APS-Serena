```python
import numpy as np
import matplotlib.pyplot as plt

def mi_funcion_sen(vmax=1, dc=0, ff=1, ph=0, nn=1000, fs=80000):
    
    t = np.arange(0, nn) / fs
    x = dc + vmax * np.sin(2 * np.pi * ff * t + ph)
    return t, x

N = 1000
fs = 80000
tt, xx = mi_funcion_sen(vmax=1, dc=0, ff=1, ph=0, nn=N, fs=fs)
tt, m = mi_funcion_sen(vmax=0.5, dc=0, ff=2, ph=0, nn=N, fs=fs)
xx3 = m * xx

plt.plot(tt, xx,color='orchid')
plt.title("Seno parametrizado")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.grid(True)
plt.show()

plt.plot(tt, xx3,color='orange')
plt.title("Seno parametrizado")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.grid(True)
plt.show()

plt.plot(tt, m,color='blue')
plt.title("Seno parametrizado")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.grid(True)
plt.show()
```
