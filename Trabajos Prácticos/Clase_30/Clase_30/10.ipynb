```python
from scipy.ndimage import median_filter
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
fs = 1000  # Frecuencia de muestreo en Hz
ventana_200ms = int(0.2 * fs)  # Ventana de 200 ms
ventana_600ms = int(0.6 * fs)  # Ventana de 600 ms

# Seal de ejemplo
s = np.random.randn(50000)  # Seal simulada

# Aplicar filtro de mediana de 200 ms
m = median_filter(s, size=ventana_200ms)

# Aplicar filtro de mediana de 600 ms sobre el resultado
b_estimado = median_filter(m, size=ventana_600ms)

print("Filtro de mediana aplicado correctamente con median_filter.")

# Graficar la seal original
plt.figure(figsize=(10, 6))
plt.plot(s, label='Seal original', alpha=0.7)
plt.title('Seal Original')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.legend()
plt.grid()
plt.show()

# Graficar la seal despus del filtro de mediana de 200 ms
plt.figure(figsize=(10, 6))
plt.plot(m, label='Seal filtrada (200 ms)', alpha=0.7, color='orange')
plt.title('Seal Filtrada (200 ms)')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.legend()
plt.grid()
plt.show()

# Graficar la seal despus del filtro de mediana de 600 ms
plt.figure(figsize=(10, 6))
plt.plot(b_estimado, label='Seal filtrada (600 ms)', alpha=0.7, color='green')
plt.title('Seal Filtrada (600 ms)')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.ndimage import median_filter
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
fs_ecg = 1000  # Frecuencia de muestreo en Hz
soporte_200ms = int(0.2 * fs_ecg)  # Ventana de 200 ms
soporte_600ms = int(0.6 * fs_ecg)  # Ventana de 600 ms

# Cargar la seal desde el archivo .npy
ecg = np.load('ecg_sin_ruido.npy')  # Cargar seal ECG sin ruido
N_ecg = len(ecg)  # Tamao de la seal

# FILTROS DE MEDIANA
# Filtro de mediana de 200 ms
m = median_filter(ecg, size=soporte_200ms)
# Filtro de mediana de 600 ms sobre el resultado
b_estimado = median_filter(m, size=soporte_600ms)

print("Filtro de mediana aplicado correctamente.")

# Graficar la seal original
plt.figure(figsize=(10, 6))
plt.plot(ecg, label='Seal original', alpha=0.7)
plt.title('Seal Original')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.legend()
plt.grid()
plt.show()

# Graficar la seal despus del filtro de mediana de 200 ms
plt.figure(figsize=(10, 6))
plt.plot(m, label='Seal filtrada (200 ms)', alpha=0.7, color='orange')
plt.title('Seal Filtrada (200 ms)')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.legend()
plt.grid()
plt.show()

# Graficar la seal despus del filtro de mediana de 600 ms
plt.figure(figsize=(10, 6))
plt.plot(b_estimado, label='Seal filtrada (600 ms)', alpha=0.7, color='green')
plt.title('Seal Filtrada (600 ms)')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
import scipy.io as sio


#ECG
mat_struct = sio.loadmat('./ECG_TP4.mat')

ecg_one_lead = mat_struct['ecg_lead'].flatten()
N = len(ecg_one_lead)
cant_muestras = N

#cantidad ded muestras de la mediana
cant_muestras_med = 201

ECG_med200 = signal.medfilt(ecg_one_lead, cant_muestras_med) #kernel size impar porque algo par no tiene media
#200 es la canti

ECG_med600 = signal.medfilt(ECG_med200, 601) #kernel size impar porque algo par no tiene media

resta = ecg_one_lead-ECG_med600

plt.figure(1)
plt.plot(ecg_one_lead, label = 'ecg', color = 'orchid')
plt.plot(ECG_med600, label = 'med 600', color = 'cornflowerblue')
plt.plot(resta, label = 'resta', color = 'rebeccapurple')
# plt.title('Respuesta en Magnitud')
# plt.xlabel('Frecuencia [Hz]')
# plt.ylabel('|H(j)| [dB]')
plt.legend()
plt.grid(True, which='both', ls=':')
```
