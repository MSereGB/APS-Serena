## 1(a) Señal sinusoidal de 2 kHz

Defino $x_1(t) = A \sin(2\pi f t)$ con $A=1$ y $f=2\,\text{kHz}$.
Indico el **tiempo entre muestras** ($\Delta t$), el **número de muestras** ($N$) y la **potencia promedio** en la ventana, y grafico un recorte temporal.



```python
import numpy as np
import matplotlib.pyplot as plt

# 1(a) — Seno 2 kHz (N=200, T=10 ms)
N = 200
T = 0.010
t = np.linspace(0, T, N, endpoint=False)
Ts = T / N

x = np.sin(2*np.pi*2000*t)
P = (x**2).mean()

print(f"Δt = {Ts:.6e} s")
print(f"N  = {N}")
print(f"Potencia = {P:.6f}")

idx = (t >= 0) & (t <= 2e-3)
plt.plot(t[idx], x[idx])
plt.title('1(a) Seno 2 kHz — ventana 2 ms')
plt.xlabel('Tiempo (s)'); plt.ylabel('Amplitud'); plt.grid(True)
plt.show()
```
