```python
import numpy as np
import matplotlib.pyplot as plt

n= np.arange(0, 6*np.pi,0.2) #secuencia de puntos parte de la funcion evaluados en la funcion seno
fun = np.sin(n)


plt.plot(n, fun, color = 'red', marker = '*')
plt.title("Un periodo del seno")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
